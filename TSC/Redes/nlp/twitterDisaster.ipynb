```python
# Cleaning tweets (keeping hastags, removing mentions, urls...)
import preprocessor as twt_p
twt_p.set_options(twt_p.OPT.URL, 
                  twt_p.OPT.EMOJI,
                  twt_p.OPT.SMILEY,
                  twt_p.OPT.NUMBER,
                  twt_p.OPT.RESERVED,
                  twt_p.OPT.MENTION)

twt_p.clean(trn_data.text[0])
```

```python
for i,v in enumerate(trn_data.text):
    trn_data.loc[i,"text"] = twt_p.clean(v)

for i,v in enumerate(tst_data.text):
    tst_data.loc[i,"text"] = twt_p.clean(v)

trn_data.head()
```

```python
# Remove numbers
trn_data.text = trn_data.text.astype(str).str.replace('\d+', '')

# To lower case
trn_data.text = trn_data.text.str.lower()

trn_data.text
```

```python
# From: https://towardsdatascience.com/basic-tweet-preprocessing-in-python-efd8360d529e
# Lemmatize, remove stop words and punctuation

import re
import nltk
from nltk.tokenize import TweetTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('wordnet')

lemmatizer = nltk.stem.WordNetLemmatizer()
w_tokenizer = TweetTokenizer()


def lemmatize_text(text):
  return [(lemmatizer.lemmatize(w)) for w in w_tokenizer.tokenize((text))]

def remove_punctuation(words):
  new_words = []
  for word in words:
      new_word = re.sub(r'[^@\w\s]', '', (word))
      if new_word != '':
        new_words.append(new_word)
  return new_words


trn_data.text = trn_data.text.apply(lemmatize_text)
trn_data.text = trn_data.text.apply(remove_punctuation)

stop_words = set(stopwords.words('english'))

trn_data.text = trn_data.text.apply(lambda x: [item for item in x if item not in stop_words])
trn_data.text
```

Para representar los mensajes sin perder el orden, creamos una vectorización por índices

```python
# Create vectorization by count of the tweets
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

top_words = 10000

t = Tokenizer(num_words=top_words)
t.fit_on_texts(trn_data.text)

# vectorized_trn = t.texts_to_matrix(trn_data.text, mode='count')
vectorized_trn = t.texts_to_sequences(trn_data.text)


# The maximum number of words is in the longest list
def find_max_list(list):
    list_len = [len(i) for i in list]
    max_len = max(list_len)

    print("Longest tweet after preprocessing: {}".format(max_len))

    return max_len


max_words = find_max_list(vectorized_trn)
max_words = 15
vectorized_trn = pad_sequences(vectorized_trn, max_words)

vectorized_trn
```

```python
# Train-test (local) split
from sklearn.model_selection import train_test_split

x_trn, x_tst, y_trn, y_tst = train_test_split(vectorized_trn, trn_data.target, test_size=0.2)

x_trn
```

```python
################################################################################
# Define layers
################################################################################

from keras.models import Sequential
from keras.layers import Dense,Flatten,Dropout
from keras.layers import LSTM
from keras.layers import SimpleRNN
from keras.layers.embeddings import Embedding
from keras.optimizers import Adam

from keras.callbacks import EarlyStopping

from sklearn.metrics import confusion_matrix
```

```python
################################################################################
# Test model
################################################################################

def plot_history(history):
  acc = history.history["accuracy"]
  val_acc = history.history["val_accuracy"]

  loss = history.history["loss"]
  val_loss = history.history["val_loss"]

  epochs = range(1, len(acc)+1)

  plt.plot(epochs, acc, "r", label="Training acc")
  plt.plot(epochs, val_acc, "b", label="Validation acc")
  plt.title("Training and validation accuracy")
  plt.xlabel("Epoch")
  plt.ylabel("Accuracy")
  plt.legend()
  plt.figure()

  plt.plot(epochs, loss, "r", label="Training loss")
  plt.plot(epochs, val_loss, "b", label="Validation loss")
  plt.title("Training and validation loss")
  plt.xlabel("Epoch")
  plt.ylabel("Loss")
  plt.legend()
  plt.figure()

  plt.show()

# ------------------------------------------------------------------------------

def evaluate_model(model, history):
  print(model.summary())

  print("Max words:  {}".format(max_words))
  print("Max Epochs: {}".format(epochs))
  print("Last epoch: {}".format(len(history.epoch)))
  print("Batch size: {}".format(batch_size))
  print("Learning r: {}".format(lr))

  loss, acc = model.evaluate(x_tst, y_tst, verbose=0)  
  print("\nTest loss: {}".format(loss))
  print("Accuracy:  {}\n".format(acc))

  pred = model.predict(x_tst)
  pred = (pred>.5)
  cm = confusion_matrix(y_tst, pred)

  print(pd.DataFrame(cm))

  plot_history(history)
```

```python
################################################################################
# Fit model
# FEEDFORWARD MODEL
################################################################################

# Define model
model = Sequential([
  Embedding(top_words, 100, input_length=max_words),
  Flatten(),
  Dense(16, activation="relu"),
  Dropout(0.2),
  Dense(16, activation="relu"),
  Dropout(0.2),
  Dense(1, activation="sigmoid")
])

# ------------------------------------------------------------------------------

lr = 5e-5

model.compile(
    loss="binary_crossentropy",
    # optimizer="rmsprop",
    # optimizer="adam",
    optimizer=Adam(learning_rate=lr),
    metrics=["accuracy"]
)

# ------------------------------------------------------------------------------

# Parameters
epochs = 50
batch_size = 105

# Train model
history = model.fit(
              x_trn, y_trn,              
              validation_split=0.2,
              epochs=epochs,
              batch_size=batch_size,
              callbacks=[EarlyStopping(monitor='val_accuracy', 
                                       patience=5)],
              verbose=0
          )
```
